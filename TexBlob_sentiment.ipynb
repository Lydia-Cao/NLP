```python
string1 = "ITEM 7"
string2 = "ITEM 7A"
f1 = open("2016.txt",encoding = "latin1")
index = 0
index_list = []
end = 0
end_list = []
for line in f1:
    index += 1
    if string1 in line:
        print(('String', string1, 'Found In Line', index))
        index_list.append(index)
        continue
f2 = open("2016.txt", encoding = "latin1")
for l in f2:
    end += 1
    if string2 in l:
        print(('String', string2, 'Found In Line', end))
        end_list.append(end)
        continue
start = index_list[0]
end1 = end_list[0]
content = range(start, end1)

f3 = open('2016.txt', encoding = "latin1")
lines = []
for pos, l in enumerate(f3):
    if pos in content:
        lines.append(l)
    continue

new = []
for l in lines:
    new_line = l.split("\n")
    new += new_line

non_empty_lines = [n for n in new if
                   n.strip() != "" and n.strip() != "------------------------------------------------------------------------" and n.strip() != "Table of Contents <#toc>"]
string_without_empty_lines = ""
for line in non_empty_lines:
    string_without_empty_lines += line + "\n"
string_without_empty_lines = string_without_empty_lines.strip("\n").strip("n").strip("\t")
print(string_without_empty_lines)

import nltk
from textblob import TextBlob

words = [TextBlob(c.lower()) for c in list(string_without_empty_lines.split("."))]
Sentiment = [t.sentiment.polarity for t in words]
po = sum(Sentiment)/len(Sentiment)
print(po)

f1.close()
f2.close()
f3.close()
```
