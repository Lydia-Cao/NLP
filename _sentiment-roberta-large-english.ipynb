```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification 

tokenizer = AutoTokenizer.from_pretrained("siebert/sentiment-roberta-large-english") 

model = AutoModelForSequenceClassification.from_pretrained("siebert/sentiment-roberta-large-english")
```

```python
string1 = "Management's"
string2 = "Item 7A"
f1 = open("2021.txt",encoding = "latin1")
index = 0
index_list = []
end = 0
end_list = []
for line in f1:
    index += 1
    if string1 in line:
        print(('String', string1, 'Found In Line', index))
        index_list.append(index)
        continue
f2 = open("2021.txt", encoding = "latin1")
for l in f2:
    end += 1
    if string2 in l:
        print(('String', string2, 'Found In Line', end))
        end_list.append(end)
        continue
start = index_list[0]
end1 = end_list[0]
content = range(start, end1)

f3 = open('2021.txt', encoding = "latin1")
lines = []
for pos, l in enumerate(f3):
    if pos in content:
        lines.append(l)
    continue

new = []
for l in lines:
    new_line = l.split("\n")
    new += new_line

non_empty_lines = [n for n in new if
                   n.strip() != "" and n.strip() != "------------------------------------------------------------------------" and n.strip() != "Table of Contents <#toc>"]
string_without_empty_lines = ""
for line in non_empty_lines:
    string_without_empty_lines += line + "\n"
string_without_empty_lines = string_without_empty_lines.strip("\n").strip("n").strip("\t")
print(string_without_empty_lines)
from transformers import pipeline
sentiment_analysis = pipeline("sentiment-analysis",model="siebert/sentiment-roberta-large-english")
print(len(string_without_empty_lines))
a = []
for i in range (1,65):
    output = sentiment_analysis(string_without_empty_lines[600*i-600:600*i])
    a.append(output[0]["score"])
score = sum(a)/len(a)
print(score)
# print(sentiment_analysis(string_without_empty_lines[1:2300]))
```
